Розрахуйте коефієнти розкладу біному Ньютона за допомогою генератора (https://uk.wikipedia.org/wiki/%D0%91%D1%96%D0%BD%D0%BE%D0%BC_%D0%9D%D1%8C%D1%8E%D1%82%D0%BE%D0%BD%D0%B0).

1. Степінь многочлена користувач має вводити з клавіатури (невід'ємне ціле число).

2. Програма має підраховувати біноміальні коефіцієнти для біномів усіх степенів, починаючи від 0 і закінчуючи тим, який ввів користувач.

3. Коефіцієнти мають виводитись у вигляді трикутника Паскаля.

Наприклад, якщо користувач введе число 5, програма має вивести результат:

1

1 1

1 2 1

1 3 3 1

1 4 6 4 1

1 5 10 10 5 1


```
import math

def kof(n):
    k=0
    d = ""
    while k<=n:
        C = math.factorial(n)/(math.factorial(k)*math.factorial(n-k))
        d += str(round(C))
        k += 1
    yield d
    return 

n = int(input("Enter n "))
if n>=0:
    for i in range(0, n+1):
        a = kof(i)
        for g in kof(i):
            print(g)
else:
    print("Please enter positive n")
```

```
def register(active=True):
    def decorator(funk):
        def inner(p, n):
            print('returned from', funk.__name__)
            print('calling', funk.__name__, 'with', p, 'and', n)
            funk(p, n)
        return inner
    return decorator

@register(active=False)
def book(p, n):
    d=int(p/n)
    print("There are", d, "notebooks")
    l=[]
    z = 0
    t = n
    h=t/2
    while d >0:
        j = range(z, t + 1)
        for i in range (0, int(h)):
            l.append(j[-i-1])
            l.append(j[i + 1])
        z+=n
        t+=n
        d-=1
        print(l)
        l=[]
    return

@register()
def book1(p, n):
    d=int(p/n)
    print("There are", d, "notebooks")
    l=[]
    list=[]
    z = 0
    t = n
    h=t/2
    while d >0:
        j = range(z, t + 1)
        for i in range (0, int(h)):
            l.append(j[-i-1])
            l.append(j[i + 1])
        z+=n
        t+=n
        d-=1
        list.append(l)
        l=[]
    print(list)
    return

while True:
    try:
        a = int(input("Enter the number of pages to 1280: "))
        if a == 1280:
           b = int(input("Enter the number of pages in each notebook 16 or 24: "))
        else:
            b = int(input("Enter the number of pages in each notebook 16 or 24 or 32: "))
        c = int(input("Enter 0 or 1: "))
        if a>0 and a<=1280 and (b==16 or b==24 or  b==32) and (c==1 or c==0) and a%b==0:
            if c==0:
                book(a,b)
            else:
                book1(a,b)
        else:
            print("Entered values are incorrect!")
    except:
        print("Entered values are incorrect")
    break
```
